在二维空间中，一个贝塞尔曲线的具体输入是一组控制点，通常至少需要三个控制点来定义一条曲线。这些控制点可以用二维笛卡尔坐标系中的点来表示。

假设我们有一个三次贝塞尔曲线，它由四个控制点 $P_0$, $P_1$, $P_2$, $P_3$ 组成。这条曲线的起点是 $P_0$，终点是 $P_3$。控制点 $P_1$ 和 $P_2$ 决定了曲线在中间的弯曲程度和方向。

贝塞尔曲线的参数方程可以表示为：

$$
B(t) = (1-t)^3P_0 + 3t(1-t)^2P_1 + 3t^2(1-t)P_2 + t^3P_3
$$

其中 $t$ 是参数，它的取值范围通常是 $0\leq t\leq 1$。当 $t=0$ 时，曲线的点位于起点 $P_0$ 处；当 $t=1$ 时，曲线的点位于终点 $P_3$ 处。在 $0<t<1$ 的范围内，曲线的点位于控制点 $P_0$ 和 $P_3$ 之间，但是其实际位置和形状由控制点 $P_1$ 和 $P_2$ 的位置决定。

通过调整控制点的位置可以改变曲线的形状和弯曲程度。在计算机图形学中，通常使用图形界面工具或代码来创建和编辑贝塞尔曲线，并将其用于绘制二维图形、动画和其他视觉效果。
